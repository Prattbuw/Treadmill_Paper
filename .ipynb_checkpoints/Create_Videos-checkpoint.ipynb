### Video 1 - example treadmill walking

```python
# load data
genotype = 'wt_berlin'#'iav_kir'#'wt_berlin' 
results_dir = 'C:/Users/pratt/Desktop/Analysis/LINEAR_BELT_RESULTS/results/'
meta_matrix = np.load(results_dir + genotype + '_meta_matrix.npy', allow_pickle=True)
heading_angle = np.load(results_dir + genotype + '_heading_angle.npy', allow_pickle=True)
body_positions = np.load(results_dir + genotype + '_body_positions.npy', allow_pickle=True)
total_velocity = np.load(results_dir + genotype + '_total_velocity.npy', allow_pickle=True)
```

```python
# isolate example trial data
FS = 180
ex_idx = np.logical_and(meta_matrix[:,0] == 2, meta_matrix[:,1] == 17)
ex_BL = meta_matrix[ex_idx,-1]
ex_heading = heading_angle[ex_idx,:][0]
ex_vel = total_velocity[ex_idx,:][0]
ex_vel[0] = ex_vel[1]
ex_tx = body_positions[3,:,ex_idx][0] * ex_BL
t = np.arange(0,len(ex_tx)/FS, 1/FS) # time
belt_stim = np.zeros(len(ex_heading))
adjusted_driving_speeds = np.unique(meta_matrix[:,2])
belt_stim[0:2*FS] = np.linspace(adjusted_driving_speeds[0], adjusted_driving_speeds[2], 2*FS)
belt_stim[2*FS::] = adjusted_driving_speeds[2]
```

```python
# load in video data
vid_dir = 'H:/.shortcut-targets-by-id/10pxdlRXtzFB-abwDGi0jOGOFFNm3pmFK/Tuthill Lab Shared/manuscripts/treadmill_2023/videos/video 5 wt berlin fly example/'
vid_1 = 'fly2_Speed_15_Trial_17Camera1.avi'
frame_store_1 = []

cap = cv2.VideoCapture(vid_dir+vid_1)
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    frame_store_1.append(frame[:,10:750,:])
    
# When everything done, release 
# the video capture object 
cap.release() 
  
# Closes all the frames 
cv2.destroyAllWindows()

vid_2 = 'fly2_Speed_15_Trial_17Camera5.avi'
frame_store_2 = []

cap = cv2.VideoCapture(vid_dir+vid_2)
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    frame_store_2.append(frame[:,150:550,:])
    
# When everything done, release 
# the video capture object 
cap.release() 
  
# Closes all the frames 
cv2.destroyAllWindows()
```

```python
# plot everything
fig = plt.figure(1, figsize = (9,6))
# vid 1
plt.subplot(6,2,(1,3))
plt.imshow(frame_store_1[1000])
plt.xticks([])
plt.yticks([])

#vid2
plt.subplot(6,2,(2,4))
plt.imshow(frame_store_2[1000])
# plt.imshow(np.transpose(frame_store_2[1000],(1,0,2)))
plt.xticks([])
plt.yticks([])

plt.subplot(6,2,(5,6))
plt.plot(t, ex_tx, color ='black', linewidth = 3)
plt.yticks(fontsize = 12, fontname = 'Arial')
plt.ylabel('chamber position (mm)', fontsize = 10, fontname = 'arial')
plt.xticks([])
plt.axvspan(1428/FS, 1614/FS, facecolor='powderblue', alpha=0.5)
```

```python
frame = 100
# position
plt.subplot(4,1,1)
plt.plot(t, ex_tx, color ='black', linewidth = 3)

for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False)

plt.tick_params(axis='y', which='both', right=False, 
                left=False, labelleft=False) 

plt.tick_params(axis='x', which='both', bottom=False, 
                top=False, labelbottom=False) 

plt.xlim(0,10)

plt.subplot(4,1,1)
plt.plot(np.array(t[frame]), np.array(ex_tx[frame]), color ='red', marker ='.', markersize = 20, linestyle ='none', alpha = 0.5)
```

```python
# plot everything
# Define the meta data for the movie
FFMpegWriter = manimation.writers['ffmpeg']
metadata = dict(title='video1', artist='Matplotlib')
writer = FFMpegWriter(fps=FS, metadata=metadata)


fig = plt.figure(1, figsize = (12,8))

# body position
plt.subplot(4,1,1)
red_circle1, = plt.plot([], [], 'ro', markersize = 12, alpha = 0.7)
plt.plot(t, ex_tx, color ='black', linewidth = 3)
for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False)

plt.tick_params(axis='y', which='both', right=False, 
                left=False, labelleft=False) 

plt.tick_params(axis='x', which='both', bottom=False, 
                top=False, labelbottom=False) 

plt.xlim(0,10)

# heading
plt.subplot(4,1,2)
red_circle2, = plt.plot([], [], 'ro', markersize = 12, alpha = 0.7)
plt.plot(t, ex_heading, color ='black', linewidth = 3)
for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False)

plt.tick_params(axis='y', which='both', right=False, 
                left=False, labelleft=False) 

plt.tick_params(axis='x', which='both', bottom=False, 
                top=False, labelbottom=False) 

plt.xlim(0,10)


# vel
plt.subplot(4,1,3)
red_circle3, = plt.plot([], [], 'ro', markersize = 12, alpha = 0.7)
plt.plot(t,ex_vel, color ='black', linewidth = 3)
for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False)

plt.tick_params(axis='y', which='both', right=False, 
                left=False, labelleft=False) 

plt.tick_params(axis='x', which='both', bottom=False, 
                top=False, labelbottom=False) 

plt.xlim(0,10)

# stim
plt.subplot(4,1,4)
plt.plot(t, belt_stim, color ='black', linewidth = 3)
for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False)

plt.tick_params(axis='y', which='both', right=False, 
                left=False, labelleft=False) 

plt.tick_params(axis='x', which='both', bottom=False, 
                top=False, labelbottom=False) 

plt.xlim(0,10)

# define redcircles to update
red_circle4, = plt.plot([], [], 'ro', markersize = 12, alpha = 0.7)

with writer.saving(fig, vid_dir+"vid1.avi", 100):
    for frame in tqdm(range(len(ex_tx))):

        # position
        plt.subplot(4,1,1)
        red_circle1.set_data(np.array(t[frame]), np.array(ex_tx[frame]))
        
        # heading
        plt.subplot(4,1,2)
        red_circle2.set_data(np.array(t[frame]), np.array(ex_heading[frame]))
        
        # vel
        plt.subplot(4,1,3)
        red_circle3.set_data(np.array(t[frame]), np.array(ex_vel[frame]))
     
        # stim
        plt.subplot(4,1,4)
        red_circle4.set_data(np.array(t[frame]), np.array(belt_stim[frame]))
        
        writer.grab_frame()
```
